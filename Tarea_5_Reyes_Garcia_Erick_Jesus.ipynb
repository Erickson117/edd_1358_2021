```
"""
Nombre: Erick Jesus Reyes Garcia
Grupo: 1358
Asignatura: Estructura de datos
"""
class NodoDoble:
  def __init__( self , value , previo = None , siguiente = None ):
    self.data = value
    self.prev = previo
    self.next = siguiente
  
class DoubleLinkedList:
  def __init__(self):
    self.__head = None
    self.__tail = None
    self.__size = 0

  def get_size(self):
    return self.__size

  def is_empty(self):
    return self.__size == 0

  def append(self , value):
    if self.is_empty() :
      nuevo = NodoDoble(value)
      self.__head = nuevo
      self.__tail = nuevo
    else:
      nuevo = NodoDoble(value, self.__tail, None)
      self.__tail.next = nuevo
      self.__tail = nuevo

    self.__size += 1  

  def transversal(self):
    curr_node = self.__head
    while curr_node != None:
      print("<=== ", curr_node.data , "===>", end= "")
      curr_node = curr_node.next
    print("")

  def reverse_transversal(self): # es lo mismo que reverse solo cambie la posicion de curr_node y en lugra de usar next se usa prev
    curr_node = self.__tail
    while curr_node != None:
      print("<=== ", curr_node.data , "===>", end= "")
      curr_node = curr_node.prev #cambio de next por prev
    print("")

  def remove_from_head(self,value):
    if self.is_empty(): #caso en no haya nada
      pass
    else:
      curr_node = self.__head
      while curr_node.data != value and curr_node != None:
        curr_node = curr_node.next

      if curr_node.data == value: # caso en el que head sea el valor 
        if curr_node != self.__head and curr_node != self.__tail: #verifica si no apunta a tail o head ya que de ser asi no modificara hacia donde apuntan estos
          curr_node.prev.next = curr_node.next
          curr_node.next.prev = curr_node.prev
        else:  
          if curr_node == self.__head: #si curr_node apunta a lo que apunta head hay que volverlo nulo y hay que hacer que head apunte a el siguiente nodo
            self.__head = self.__head.next
            self.__head.prev = None
            curr_node.next = None
            curr_node = None
          else: #en caso de que apunte a tail hacemos que el valor en tail sea nulo y que tail apunte al nodo anterior
            curr_node.prev.next = None
            self.__tail = curr_node.prev
      self.__size -= 1

  def remove_from_tail(self, value): #funciona igual que "remove" pero aqui se cuenta en reversa y se cambian algunos detalles
    if self.is_empty():
      pass
    else:
      curr_node = self.__tail #en lugar de asignar head se asigna tail
      while curr_node != None and curr_node.data != value:
        
        print(curr_node.data)
        curr_node = curr_node.prev #en lugar de next se usa prev

      if curr_node.data == value:
        
        if curr_node != self.__head and curr_node != self.__tail: #funciona igual
          curr_node.prev.next = curr_node.next
          curr_node.next.prev = curr_node.prev
        else:
          if curr_node == self.__tail: # si curr_node apunta a tail, tail debe apuntar a curr_node.prev y curr_node debe ser nulo 
            curr_node.prev.next = None
            self.__tail = curr_node.prev
            curr_node = None
            
          else: #si el valor es head head debe apuntar al siguiente nod y curr_node debe ser nulo
            self.__head = curr_node.next
            curr_node = None
      self.__size -= 1

  def find_from_tail(self, value): # cuenta desde el cero la posicion en la que esta el valor dado
    curr_node = self.__tail
    posicion = self.__size -1 # se le resta uno debido a que contara desde el cero entonces si son 10 elementos contara del 0 al 9
    while curr_node != None and curr_node.data != value:
        curr_node = curr_node.prev
        posicion -= 1 #al contar desde tail va a ir en orden decendente
    return posicion

  def find_from_head(self, value):
    curr_node = self.__head
    posicion = 0 #se inicia en 0 la variable con la que obtiene el valor de la posicion
    while curr_node != None and curr_node.data != value:
        curr_node = curr_node.prev
        posicion += 1 #orden acendente
    return posicion
```
